```python
# Hausman test for fixed versus random effects model
# null hypothesis: random effects model
#calcula las matrices de covarianza de los coeficientes en ambos modelos y las resta
psi = efect_fijos_res.cov - efect_aleato_res.cov.iloc[1:,1:]
#calcula un vector con las diferencias de los coeficientes de ambos modelos
diff = efect_fijos_res.params - efect_aleato_res.params[1:]
diff = np.array(diff)[:, np.newaxis] #agregar la dimesión de 1 matriz (nx1)
diff_t= np.transpose(diff)
# psi = fixed1.cov.iloc[1:,1:] - random.cov.iloc[1:,1:]
# diff = fixed1.params[1:] - random.params[1:] #cuando el efecto fijo lleva constante
W = diff_t.dot(np.linalg.inv(psi)).dot(diff) #.dot para el poducto escalar de dos matrices
#linalg.inv(psi) calcula la inversa multiplicativa de la matriz psi
#entonces se multiplica la matriz diff por la inversa multiplicativa de la matriz psi, lo cual genera otra matriz
#grados de libertad
dof = efect_aleato_res.params.size -1
pvalue = stats.chi2(dof).sf(W) #sf es la función de supervivencia, igual a 1- F(x) 1- la función de distribución acumulada
print("Hausman Test: chisq = {0}, df = {1}, p-value = {2}".format(W, dof, pvalue))
```

```python
#table = dp.Table(df8_mod)
# markdown_1 = dp.Markdown("""# Balanced panel database. Active Economically Population (PEA in Spanish). Colombia, 2010-2016""")

# table_2 = dp.Table(dfg_mod) 
# markdown_2 = dp.Markdown("""# Balanced panel database. Treat and control groups. Colombia, 2010-2016""")

#subir a datapane las bases de datos y el código:
desc = ''' 
Panel de datos balanceado con las observaciones correspondientes a las unidades transversales (personas) 
y temporales (años) para los grupos de tratamiento y control provenientes del trabajo "Efecto Scarring del
desempleo sobre las remuneraciones y la calidad del empleo en la zona urbana. Colombia, 2010 – 2016" (pendiente de publicación).
Esta tabla se llama dfg_mod (data frame groups model):'''
table_2 = dp.DataTable (dfg_mod)
desc_t2 = ''' 
Panel de datos construido para correr los modelos de probabilidad logit y probit para panel de datos en stata 17, ya que no están disponibles en python a la fecha del 19-04-2022:
'''
table_3 = dp.DataTable(dfg_mod_c_sta)

desc_code = ''' 
Código utilizado para correr los modelos de panel de datos de efectos fijos, aleatorios y primeras diferencias, con variable dependiente ingreso (y) y variables explicativas (x):
'''
code = '''
import pandas as pd
import numpy as np
import statsmodels.api as sm 
import linearmodels as lm

#Cargar la base dfg_mod. Descargan la tabla dfg_mod y colocan la ruta de acceso en la siguiente línea:
dfg_mod= pd.read_csv('path/name_file.csv', sep = ";",  index_col=[0,1], low_memory=False)

# Se convierte la variable dependiente a logaritmo
y = np.log(dfg_mod["Ingreso_td_hora"])

# Se extraen las variables independientes
x = dfg_mod[["sexo","scar2", "region_encuesta","edad", "estado_civil", "vive_conyuge", 
              "educación3", "descripcion_ciiu2", "exper", "Tipo_empleo2",
             "ano"]]
# Se crean variables al cuadrado
x["edad2"] = x["edad"]**2
x["exper2"] = x["exper"]**2

#Convertimos todas las variables categóricas a dummy, eliminando una categoria, la cual es la base de comparación.
#Extraemos el subconjunto de dichas variables. 
x = pd.get_dummies(x,columns=["sexo","region_encuesta","estado_civil", "vive_conyuge", 
                              "educación3", "descripcion_ciiu2", 
                              "Tipo_empleo2", "ano"], drop_first=True)
#interacciónes categóricas:
x["sexo_its_mujer"] = x["sexo_mujer"]*x["scar2"] 
x["ed_Univer_its"] = x["educación3_Universitario"]*x["scar2"]
x["ed_pos_its"] = x["educación3_Posgrado"]*x["scar2"]
#Se debe habilitar esta linea para correr el modelo de efectos fijos.
#x.drop(columns=["sexo_mujer"], inplace=True)

w = dfg_mod["factorex_plong"]

# Econometrics Models Income Scar. 
# Modelo de efectos fijos
efect_fijos = lm.PanelOLS(y, x,  weights = w, entity_effects=True, time_effects=False) #sm.add_constant(x)
efect_fijos_res = efect_fijos.fit()#cov_type="robust"
efect_fijos_res.summary

# Modelo de primeras diferencias
pmra_dif = lm.FirstDifferenceOLS(y, x, weights = w) #sm.add_constant(x)
pmra_dif_res = pmra_dif.fit(cov_type='robust') 
pmra_dif_res.summary

# Modelo de efectos aleatorios
efect_aleato = lm.RandomEffects(y, sm.add_constant(x),  weights = w)
efect_aleato_res = efect_aleato.fit()#cov_type='robust'
efect_aleato_res.summary

#Comparar los  modelos en una sola tabla
from linearmodels.panel import compare
# las claves del diccionario son los nombres de los modelos con sus respectivas estancias de resultados

dicc = {"P. Dif":pmra_dif_res,
        "E.Fijos":efect_fijos_res,
        "E. Aleatorios":efect_aleato_res}
        
compare(dicc, precision = "std-errors", stars=True)

# Hausman test for fixed versus random effects model
from scipy import stats
# null hypothesis: random effects model
#calcula las matrices de covarianza de los coeficientes en ambos modelos y las resta
psi = efect_fijos_res.cov - efect_aleato_res.cov.iloc[1:,1:]
# psi = fixed1.cov.iloc[1:,1:] - random.cov.iloc[1:,1:] #cuando el efecto fijo lleva constante
#calcula un vector con las diferencias de los coeficientes de ambos modelos
diff = efect_fijos_res.params - efect_aleato_res.params[1:]
# diff = fixed1.params[1:] - random.params[1:] #cuando el efecto fijo lleva constante
diff = np.array(diff)[:, np.newaxis] #agregar la dimesión de 1 matriz (nx1)
diff_t= np.transpose(diff)

#se multiplica la matriz diff por la inversa multiplicativa de la matriz psi, lo cual genera otra matriz

W = diff_t.dot(np.linalg.inv(psi)).dot(diff) #.dot para el poducto escalar de dos matrices
#linalg.inv(psi) calcula la inversa multiplicativa de la matriz psi

#grados de libertad
dof = efect_aleato_res.params.size -1
pvalue = stats.chi2(dof).sf(W) #sf es la función de supervivencia, igual a 1- F(x) 1- la función de distribución acumulada
print("Hausman Test: chisq = {0}, df = {1}, p-value = {2}".format(W, dof, pvalue))

'''
desc_code_2 = ''' Código utilizado para correr los modelos de panel de datos de probabilidad lineal de efectos fijos, aleatorios y primeras diferencias, con variable dependiente tipo de empleo informal (y) y variables explicativas (x):
'''
code_2 = ''' 
import pandas as pd
import numpy as np
import statsmodels.api as sm 
import linearmodels as lm

#Cargar la base dfg_mod. Descargan la tabla dfg_mod y colocan la ruta de acceso en la siguiente línea:
dfg_mod= pd.read_csv('path/name_file.csv', sep = ";",  index_col=[0,1], low_memory=False)

yc = dfg_mod["Tipo_empleo2"]
yc = pd.get_dummies(yc, columns=["Tipo_empleo2"], drop_first=True)
xc = dfg_mod[["sexo","scar2","region_encuesta","edad", "estado_civil", "educación3",
              "vive_conyuge", "descripcion_ciiu2","ano"]]
xc["edad2"] = xc["edad"]**2

#por tanto convertimos todas las categóricas a dummy-1 categoria que es la de comparación.
#procesamos las variables y extraemos el subconjunto de dichas variables. 
xc = pd.get_dummies(xc,columns=["sexo","region_encuesta", "estado_civil", "educación3",
                                "vive_conyuge", "descripcion_ciiu2",
                                "ano"], drop_first=True) #
#doble interacción
xc["sexo_its"] = xc["sexo_mujer"]*xc["scar2"]

#educación interactuada
xc["ed_nin_its"] = xc["educación3_Ninguno"]*xc["scar2"]
xc["ed_sec_its"] = xc["educación3_Básica secundaria y media (6 a 13)"]*xc["scar2"]
xc["ed_tec_its"] = xc["educación3_Técnico"]*xc["scar2"]
xc["ed_tecnolo_its"] = xc["educación3_Tecnológico"]*xc["scar2"]
#educ universitario
xc["ed_Univer_its"] = xc["educación3_Universitario"]*xc["scar2"]

#educ posgrado
xc["ed_pos_its"] = xc["educación3_Posgrado"]*xc["scar2"]

xc.drop(columns=["sexo_mujer"], inplace=True) #"educación3_Universitario", "educación3_Posgrado"
wc = dfg_mod["factorex_plong"]

#efectos fijos
e_fijos_cl = lm.PanelOLS(yc, xc, weights=wc, entity_effects=True, time_effects=False) #sm.add_constant(x)
e_fijos_res_cl = e_fijos_cl.fit(cov_type="robust")#cov_type="robust" las estimaciones no alcanzan a ser significcativas
e_fijos_res_cl.summary
#primeras diferencias.
p_dif_cl = lm.FirstDifferenceOLS(yc, xc, weights = wc)
p_dif_res_cl = p_dif_cl.fit(cov_type="robust")#cov_type="robust")
p_dif_res_cl.summary
#efectos aleatorios 
efect_aleato_c = lm.RandomEffects(yc, sm.add_constant(xc))
efect_aleato_res_c = efect_aleato_c.fit(cov_type="robust")
efect_aleato_res_c.summary
#comparar los  modelos en una sola tabla
from linearmodels.panel import compare
# las claves del diccionario son los nombres de los modelos, con sus respectivas estancias de resultados
#Comparación entre grupos: tratamiento
dicc = {"P. diferencias": p_dif_res_cl,
        "E.Fijos cal":e_fijos_res_cl,
         "E.Aleatorios Cal":efect_aleato_res_c}
# Hausman test for fixed versus random effects model calidad
# null hypothesis: random effects model
psi = e_fijos_res_cl.cov - efect_aleato_res_c.cov.iloc[1:,1:]
diff = e_fijos_res_cl.params - efect_aleato_res_c.params[1:]
diff = np.array(diff)[:, np.newaxis]
diff_t= np.transpose(diff)
W = diff_t.dot(np.linalg.inv(psi)).dot(diff) 
dof = efect_aleato_res_c.params.size -1
pvalue = stats.chi2(dof).sf(W)
#pvalue = stats.chi2.sf(W, dof)
print("Hausman Test: chisq = {0}, df = {1}, p-value = {2}".format(W, dof, pvalue))
'''
dp.Report(desc, table_2, desc_code,
   dp.Code(
      code=code,
      language="python"
   ), desc_code_2,
          dp.Code(
      code=code_2,
      language="python"
   ), desc_t2, table_3
).upload(name='Balanced panel database. Econometric model of income scar and python code. Colombia, 2010-2016')
```
