# Regresión logística

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función logística (sigmoide)
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generar valores de entrada (z) desde -10 hasta 10
z = np.linspace(-10, 10, 500)
# Calcular los valores de salida (sigmoide) para cada valor de z
sigmoid_values = sigmoid(z)

# Crear la gráfica
size=18
plt.figure(figsize=(14, 6))

# Graficar la función sigmoide
plt.subplot(1, 2, 1)
plt.plot(z, sigmoid_values, color="blue")
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=1)
plt.axvline(x=0, color='red', linestyle='--', linewidth=1)
plt.xlabel("$z$", fontsize=size)
plt.ylabel("$p$", fontsize=size)
plt.title("Función logística (sigmoide)", fontsize=size)
plt.xticks(fontsize=size)
plt.yticks(fontsize=size)
# plt.legend()
plt.grid(True)

# Graficar la función -log(p) y -log(1-p)
plt.subplot(1, 2, 2)
plt.plot(sigmoid_values, -np.log(1-sigmoid_values),label='$-log(1-p)$', color="green")
plt.plot(sigmoid_values, -np.log(sigmoid_values),label='$-log(p)$', color="red")
plt.axvline(x=0.5, color='blue', linestyle='--', linewidth=1)
plt.xlabel("$p$", fontsize=size)
plt.ylabel("$-log()$", fontsize=size)
# plt.title("- Log(p)", fontsize=size)
plt.legend(loc=(0.54, 0.5), fontsize=size)
plt.grid(True)
plt.xticks(fontsize=size)
plt.yticks(fontsize=size)
plt.tight_layout()
plt.show()
```

### Límites de decisión

```python
from sklearn.datasets import load_iris

# cargar datos de iris
iris = load_iris(as_frame=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Escoger sólo ancho de pétalo como característica
X = iris.data[["petal width (cm)"]].values
# Escoger virginica como clase 1 y las demás como clase 0
y = iris.target_names[iris.target] == 'virginica'
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Entrenar modelo
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Graficar límites de decisión de acuerdo con el modelo entrenado

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0, 0]

plt.figure(figsize=(8, 3))
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2,
         label="Probabilidad NO Iris virginica")
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Probabilidad Iris virginica")
plt.plot([decision_boundary, decision_boundary], [0, 1], "k:", linewidth=2,
         label="Límite de decisión")

plt.arrow(x=decision_boundary, y=0.08, dx=-0.3, dy=0,
          head_width=0.05, head_length=0.1, fc="b", ec="b")
plt.arrow(x=decision_boundary, y=0.92, dx=0.3, dy=0,
          head_width=0.05, head_length=0.1, fc="g", ec="g")
plt.plot(X_train[y_train == 0], y_train[y_train == 0], "bs")
plt.plot(X_train[y_train == 1], y_train[y_train == 1], "g^")
plt.xlabel("Ancho del pétalo (cm)")
plt.ylabel("Probabilidad")
plt.legend(loc="center left")
plt.axis([0, 3, -0.02, 1.02])
plt.grid()

plt.show()
```

# Regresión Softmax

```python
X = iris.data[["petal length (cm)", "petal width (cm)"]].values
y = iris["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

softmax_reg = LogisticRegression(C=30, random_state=42)
softmax_reg.fit(X_train, y_train)
```

```python
from matplotlib.colors import ListedColormap

custom_cmap = ListedColormap(["#fafab0", "#9898ff", "#a0faa0"])
x0, x1 = np.meshgrid(np.linspace(0, 8, 500).reshape(-1, 1),
                     np.linspace(0, 3.5, 200).reshape(-1, 1))
# Crear una malla de puntos para graficar el fondo

X_new = np.c_[x0.ravel(), x1.ravel()]
# Convertir la malla de puntos en un arreglo bidimensional

y_proba = softmax_reg.predict_proba(X_new)
# Predecir las probabilidades de clase para cada punto en la malla

y_predict = softmax_reg.predict(X_new)
# Predecir la clase para cada punto en la malla

zz = y_predict.reshape(x0.shape)
# Darle forma similar a la malla a las predicciones de clase

plt.figure(figsize=(10, 4))
# Crear una figura de tamaño específico

plt.plot(X[y == 2, 0], X[y == 2, 1], "g^", label="Iris virginica")
plt.plot(X[y == 1, 0], X[y == 1, 1], "bs", label="Iris versicolor")
plt.plot(X[y == 0, 0], X[y == 0, 1], "yo", label="Iris setosa")
# Graficar los puntos de datos de diferentes clases con colores y marcas específicas

plt.contourf(x0, x1, zz, cmap=custom_cmap)
# Graficar el fondo con los colores predichos

plt.xlabel("Longitud del Pétalo")
plt.ylabel("Ancho del Pétalo")
plt.legend(loc="center left")
plt.axis([0.5, 7, 0, 3.5])
plt.grid()
plt.show()
```
