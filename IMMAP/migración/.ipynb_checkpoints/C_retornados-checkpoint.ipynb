#### Estandarización y creación de variables

<span style="color:yellow;">Educación </span>

```python
### 1. Acceso escolar a educación inicial básica y media
### No asiste a la escuela < 18 años
def asistencia(df):
    if (df["asistencia_eduformal"]==2 and df["edad"]<18):
        x = 1
    elif (pd.isna(df["asistencia_eduformal"]) or pd.isna(df["edad"])):
        x = np.nan
    else:
        x = 0
    return(x)
retor["no_asistencia"] = retor.apply(asistencia, axis=1)
```

```python
### 2. No asiste a educación superior en la edad para hacerlo
def asis_edusup(df):
    if (df["asistencia_eduformal"]==2 and (df["edad"]>17 and df["edad"]<29)):
        x = 1
    elif (pd.isna(df["asistencia_eduformal"]) or pd.isna(df["edad"])):
        x = np.nan
    else:
        x = 0
    return(x)
retor["asistencia_esup"] = retor.apply(asis_edusup, axis=1)
```

```python
### 3. Asistencia escolar  < 5 días
# establecer semillillas
np.random.seed(1234)
retor["asiste_menos5d"] = np.random.uniform(size=len(retor)) < 0.08126
```

```python
#para convertir a entero se deben reemplazar los valores NaN de pandas por nan numpy <NA>,
#depués se convierte en float con palabra reservada y después a entero con función pandas
retor["nivel_edu"] = retor["nivel_edu"].fillna(np.nan).astype(float).astype(pd.Int64Dtype())
#posteriormente se debe remplazar los pd.NA por -1 y el -1 por pd.nan 
```

```python
## posteriormente se usa la función pd.notna en un primer nivel:
#si no es nulo ejecuta la condición anidada, en caso contrario queda pd.NA <NA>
### 4. Extraedad
def extra(df):
    if pd.notna(df["asistencia_eduformal"]) and pd.notna(df["nivel_edu"]) and pd.notna(df["edad"]):
        # prescolar
        if (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==2 and df["edad"] >= 7):
            x = 1
        # primero primaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==3 and 
              df["años_estudio"]==1 and df["edad"] >= 8) :
            x = 1
        # segundo primaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==3 and 
              df["años_estudio"]==2 and df["edad"] >= 9) :
            x = 1
        # tercero primaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==3 and 
              df["años_estudio"]==3 and df["edad"] >= 10) :
            x = 1
        # cuarto primaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==3 and 
              df["años_estudio"]==4 and df["edad"] >= 11) :
            x = 1
        # quinto primaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==3 and 
              df["años_estudio"]==5 and df["edad"] >= 12) :
            x = 1
        # sexto secundaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==4 and 
              df["años_estudio"]==6 and df["edad"] >= 13) :
            x = 1
        # séptimo secundaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==4 and 
              df["años_estudio"]==7 and df["edad"] >= 14) :
            x = 1
        # octavo secundaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==4 and 
              df["años_estudio"]==8 and df["edad"] >= 15) :
            x = 1
        # noveno secundaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"]==4 and 
              df["años_estudio"]==9 and df["edad"] >= 16) :
            x = 1
        # décimo secundaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"] in [5,6] and 
              df["años_estudio"]==10 and df["edad"] >= 17) :
            x = 1
        # once secundaria
        elif (df["asistencia_eduformal"] == 1 and df["nivel_edu"] in [5,6] and 
              df["años_estudio"]==11 and df["edad"] >= 18) :
            x = 1
        else:
            x = 0
    else:
        x = np.nan
    return x

retor["extraedad"] = retor.apply(extra, axis=1)
```

```python
### INDICADOR AGRUPADO SALUD
# 7 SAL_ACC
def sal_acceso(df):
    if (df["sin_salud"]==1 and df["barreras_salud"]==1 and 
       df["sin_tratamiento"]==1 and  df["hsalud_sexualr"]==1):
        x = 1
    elif (pd.isna(df["sin_salud"]) or pd.isna(df["barreras_salud"]) or 
         pd.isna(df["sin_tratamiento"]) or pd.isna(df["hsalud_sexualr"])):
        x = np.nan
    else:
        x = 0
    return(x)
retor["SAL_ACC"] = retor.apply(sal_acceso, axis=1)
```

```python
# 8 SAL_CON
def salcon(df):
    if df["hsalud_mental"]==1 and df["hcontami_cerca"]==1:
        x = 1
    elif pd.isna(df["hsalud_mental"]) or pd.isna(df["hcontami_cerca"]):
        x = np.nan
    else:
        x = 0
    return(x)
retor["SAL_CON"] = retor.apply(salcon, axis=1)
```

```python
### 1. PROT_SE
def protecg(df):
    if (df["rest_transito"]==1 and df["rprotec_conocido"]==1 and 
       df["trata_conocido"]==1):
        x = 1
    elif (pd.isna(df["rest_transito"]) or pd.isna(df["rprotec_conocido"]) or 
         pd.isna(df["trata_conocido"])):
        x = np.nan
    else:
        x = 0
    return(x)
retor["PROT_SE"] = retor.apply(protecg, axis=1)
```

```python
### 2. PROT_NN
def protecnna(df):
    if (df["NNA_rprotec"]==1 and df["NNA_sin_asis_protec"]==1 and 
       df["NNA_perfil"]==1):
        x = 1
    elif (pd.isna(df["NNA_rprotec"]) or pd.isna(df["NNA_sin_asis_protec"]) or 
         pd.isna(df["NNA_perfil"])):
        x = np.nan
    else:
        x = 0
    return(x)
retor["PROT_NN"] = retor.apply(protecnna, axis=1)
```

```python
# calcula cada indicador sobre el total del universo que debe responder la pregunta, 
#para caso de la tasa de desempleo se calcula sobre la población economicamente activa

pin_ind = {} 

for i in total_var:
    not_null_mask = pd.notna(retor[i])
    
    if not_null_mask.any() and i != "desocupados":
        numerator = np.nansum(retor[i][not_null_mask] * retor['fex_corregido'][not_null_mask])
        denominator = np.nansum(retor['fex_corregido'][not_null_mask])
        
        if denominator != 0:  # para evitar dividir por cero
            value = (numerator / denominator * 100)
        else:
            value = np.nan
    
    elif  not_null_mask.any() and i == "desocupados":
        numerator = np.nansum(retor["desocupados"][not_null_mask] * retor['fex_corregido'][not_null_mask]) 
        denominator = np.nansum(retor['desocupados'] *retor['fex_corregido'][not_null_mask]) + np.nansum(retor['pob_ocupada'] *retor['fex_corregido'])
        if denominator != 0:  
            value = (numerator / denominator * 100)
        else:
            value = np.nan
    else:
        value = np.nan
        
    pin_ind[f"{i}_pin"] = value

pin_ind_df = pd.DataFrame([pin_ind])
pin_ind_df.transpose()
```

#### Cálculo del PIN global

```python
#### calculo por dimensión
# lista de variables output
dim = ['educación','alojamiento','integración','protección_agrupado',
       'salud_agrupado','transporte','wash','cari','nutri_dim']
#recorre las dimensiones
# selecciona la primera fila de cada variable de la dimensión
# las convierte en un array de 2 dimensiones. ver: np.shape(np.column_stack([retor[i] for i in j]))
# y suma cada elemento del array omitiendo nulos 
#y el resultado lo divide en el total de variables de la dimensión
for j, k in zip(total_var2, dim):
    retor[k] = np.nansum(np.column_stack([retor[i] for i in j]), axis=1) / len(j)
    #values = np.column_stack([retor[i] * retor["fex_corregido"] for i in j]) # para expnadir
```

```python
### sumatoria de (X_j/h) / H
retor["prom_dim"] = np.nansum(retor[dim], axis=1) /9
```

```python
### sumatoria de (X_ij/h) / H
# crear variable dummy con más de 3 privaciones 0.33
retor["dummy_pin"] = retor.apply(lambda x: 1 if x["prom_dim"]>0.33 else 0,axis=1)
```

#### Cálculo del PiN por dimensión

```python
#generar variables por dimensión
dim_pin = ['educación_pin','alojamiento_pin','integración_pin','protección_agrupado_pin',
       'salud_agrupado_pin','transporte_pin','wash_pin','cari_pin','nutri_dim_pin']
for j, k in zip(total_var2, dim_pin):
    retor[k] = retor.apply(lambda x: 1 if np.any(
        np.nansum(np.column_stack([x[i] for i in j]), axis=1)>0) and
                           x["dummy_pin"]>0 else 0, axis=1)  
```
