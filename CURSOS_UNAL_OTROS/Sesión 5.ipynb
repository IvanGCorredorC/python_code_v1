```python
import urllib3
import requests #busquedas en páginas web
import csv

from bs4 import BeautifulSoup


urllib3.disable_warnings()
```

```python
# Realizamos el Scraping web con Beautiful Soup    
url_page = 'https://www.superfinanciera.gov.co/inicio/informes-y-cifras/cifras/establecimientos-de-credito/informacion-periodica/mensual/cosechas/cifras/desembolsos-aprobaciones-61037'
page = requests.get(url_page,verify=False).text 
soup = BeautifulSoup(page, "lxml") #extrae archivos con a href donde esta el link que se desea descargar, html extrae noticias
contenido = soup.find('div', attrs={'class': 'pub'}) #encuentra todos los div class que son pub, solo hay uno. despues del pub, aparece el a href
items = contenido.find_all('a')
items
```

```python
requests.get(url_page, verify =False).text #Para que la consulta no almacene el ip del pc
```

```python
el_tiempo = 'https://www.eltiempo.com/mundo/elecciones-estados-unidos-2020/ee-uu-donald-trump-y-las-acciones-que-ha-llevado-a-cabo-para-deslegitimar-las-elecciones-548209'
BeautifulSoup(el_tiempo, 'lxlm')
```

```python
urls = []  # La url contiene todo el código html del link de descarga, incluido título, y <a> href ... 
links = [] # el link contiene solo el <a> href de la url 
fecha = [] # la fecha de obtiene del título del <a> de cada url 

for item in items:
    urls.append(item)
for i in urls:
    html = str(i)
    soup = BeautifulSoup(html)
    
    for a in soup.find_all('a', href=True):
        links.append(a['href'])
        
    fecha.append(BeautifulSoup.get_text(i)) #estructura del lenguaje html
```

```python
urls = pd.Series(urls)
links = pd.Series(links)
año = pd.Series(fecha).str.split(' ', expand=True).get(1)
mes = pd.Series(fecha).str.split(' ', expand=True).get(0)

extract = pd.concat([urls,links,año,mes], axis=1, keys=['urls','links','año','mes'])
extract['año'] = extract['año'].fillna('0')
```

# Puedo importar bases de datos directamente desde internet sin la necesidad de descargarlos primero en mi pc

```python
# cada elemento de "links" contiene la ruta + la extensión xls
# la ruta completa es :

"https://www.superfinanciera.gov.co" + links[0]
```

```python
l = 'https://www.superfinanciera.gov.co'
data = pd.read_excel(l+links[0],sheet_name=-1)
data['Año'] = [ año.iloc[0] for i in range(len(data))]
data['Mes'] = [ mes.iloc[0] for i in range(len(data))]
```

```python
data[(data['Año'] == '2018')&(data['Mes'] == 'Enero')].head(20)
```

```python
filas = len(data)
for link,año,mes in zip(extract.links[1:],extract.año[1:],extract.mes[1:]):
    if año > '2015':
        datasets = pd.read_excel(l+link,sheet_name=-1)
        datasets['Año'] = [año for i in range(len(datasets))]
        datasets['Mes'] = [mes for i in range (len(datasets))]
        filas += len(datasets)
        data = pd.concat([data,datasets], axis=0)
```

```python
df = data.copy()

df.drop('Unnamed: 0', axis = 1, inplace = True) #borrar columna

df['Unnamed: 1'] = df['Unnamed: 1'].fillna(method='ffill') # ffill: reemplaza los valores nulos y se reemplaza por el valor inmediatamente anterior hasta que encuentre un valor diferente a NA
df['Unnamed: 2'] = df['Unnamed: 2'].fillna(method='ffill')

df.set_index('Unnamed: 1', inplace=True,drop = True) # indexa el data frame para traer vivienda
```
