#### Imputar 2019 y 2020

##### Variables poblacionales

```python
###cargue de totales 
#cargue totales 2019-2022 
names_grl = ["Periodo", "Trimestre",  "Ocupados_T"]
t18_Tot = pd.read_excel("G:/Unidades compartidas/Mercado laboral/Datos/4_Output/ml_pob_impu_19_22.xlsx",     
                       sheet_name = "bogotá trim movil", header = 0,
                         nrows = 41, usecols = "A,B,L", names = names_grl)
##datos 2021
msk_T = t18_Tot["Periodo"]==2019
t18_19_Tot = t18_Tot[msk_T==True]
t18_19_Tot.reset_index(drop =True, inplace=True)
##datos 2022
msk_T_2 = t18_Tot["Periodo"]==2020
t18_20_Tot = t18_Tot[msk_T_2 ==True]
t18_20_Tot.reset_index(drop =True, inplace=True)
```

```python
#variables
l1 = ['Ocupados', 'No informa', 'Agricultura, ganadería, caza, silvicultura y pesca',
       'Explotación de minas y canteras', 'Industrias manufactureras',
       'Suministro de electricidad gas, agua y gestión de desechos',
       'Construcción', 'Comercio y reparación de vehículos',
       'Alojamiento y servicios de comida', 'Transporte y almacenamiento',
       'Información y comunicaciones', 'Actividades financieras y de seguros',
       'Actividades inmobiliarias',
       'Actividades profesionales, científicas, técnicas y servicios administrativos',
       'Administración pública y defensa, educación y atención de la salud humana',
       'Actividades artísticas, entretenimiento recreación y otras actividades de servicios']
```

```python
##reemplazar cero para división
t05_19t['No informa'] = t05_19t['No informa'].replace({0 : np.nan})
```

```python
### 2019_impu  = (2021 marco 2018 / 2021 marco 2005) * 2019 marco 2005
#impuar valores 2019. nan quedan nan
t05_19t_imp = t05_19t.copy()
t05_20t_imp = t05_20t.copy()
for i in (l1):           
    t05_19t_imp[i] = (t05_19t[i] /t05_19t["Ocupados"]) * (t18_19_Tot["Ocupados_T"]/1000).astype(np.double)
    t05_20t_imp[i] = ((t05_19t_imp[i] * (1 + ((t05_20t[i] - t05_19t[i]) / t05_19t[i])))).astype(np.double)
```

#### Concatenar

```python
lframe = [t05_19t_imp, t05_20t_imp, t18_21t, t18_22t]
```

```python
#print(type(t18_22t))
```

```python
ml_pob_impu_19_22_sec = pd.concat(lframe, axis = 0, join = 'outer', sort = False)
```

```python
ml_pob_impu_19_22_sec
```

```python
#poblaciones a millones enteros
for i in (l1):
    ml_pob_impu_19_22_sec[i] = ml_pob_impu_19_22_sec[i]*1000
```

```python
### marca valores imputados
def m_imp(df):
    if ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.notnull(df["Ocupados"]))): 
        x = 1
    elif ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.isnull(df["Ocupados"]))):
        x = np.nan
    else:
        x = 0
    return(x)
```

```python
ml_pob_impu_19_22_sec["valor Imputado"] = ml_pob_impu_19_22_sec.apply(m_imp, axis=1)
```
