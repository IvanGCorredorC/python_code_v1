```python
#datos 2019, 2020, 2021
cols= ["Periodo", "Trimestre", "Tamaño", "Ocupados"]
t05_19_21 = pd.read_excel("G:/Unidades compartidas/Mercado laboral/Datos/1_Raw/tema_1_martin.xlsx",                      
                        sheet_name = "GEIH marco 2005", header = 0, nrows = 51, 
                        usecols = "B:E", names = cols)
```

```python
t05_19_21["Trimestre"] = (pd.Series(t05_19_21["Trimestre"],                                
                                         dtype = "string").str.split( r'\d\s', expand = True).get(1))
```

```python
#convertir categorias a columnas
#cat = ["Unipersonal", "Entre 2 y 10 personas", "Entre 11 y 50 personas", "Más de 50 personas"]
##unipersonal
msk19 = t05_19_21["Tamaño"] == "Unipersonal"
t05_19_21_uni = t05_19_21[msk19==True]
t05_19_21_uni = t05_19_21_uni.drop(columns = "Tamaño", axis=1) 
t05_19_21_uni.reset_index(drop =True, inplace = True)
t05_19_21_uni.rename(columns = {'Ocupados': 'Unipersonal'}, inplace =True)
#"Entre 2 y 10 personas"
msk19 = t05_19_21["Tamaño"] == "Entre 2 y 10 personas"
t05_19_21_p = t05_19_21[msk19==True]
t05_19_21_p = t05_19_21_p.drop(columns = "Tamaño", axis=1) 
t05_19_21_p.reset_index(drop =True, inplace = True)
t05_19_21_p.rename(columns = {'Ocupados': 'Entre 2 y 10 personas'}, inplace =True)
# Entre 11 y 50 personas
msk19 = t05_19_21["Tamaño"] == "Entre 11 y 50 personas"
t05_19_21_me = t05_19_21[msk19==True]
t05_19_21_me = t05_19_21_me.drop(columns = "Tamaño", axis=1) 
t05_19_21_me.reset_index(drop =True, inplace = True)
t05_19_21_me.rename(columns = {'Ocupados': 'Entre 11 y 50 personas'}, inplace =True)
# Entre 11 y 50 personas
msk19 = t05_19_21["Tamaño"] == "Más de 50 personas"
t05_19_21_gr = t05_19_21[msk19==True]
t05_19_21_gr = t05_19_21_gr.drop(columns = "Tamaño", axis=1) 
t05_19_21_gr.reset_index(drop =True, inplace = True)
t05_19_21_gr.rename(columns = {'Ocupados': 'Más de 50 personas'}, inplace =True)
```

```python
#merge
t05_19_21_merg = t05_19_21_uni.merge(t05_19_21_p,on=['Periodo', 'Trimestre'],validate = '1:1'
                          ).merge(t05_19_21_me, on=['Periodo', 'Trimestre'],validate = '1:1'
                                 ).merge(t05_19_21_gr,on=['Periodo', 'Trimestre'],validate = '1:1')
t05_19_21_merg
```

```python
 ###fragmentar para 2019 -2021
    #2019
msk19 = t05_19_21_merg["Periodo"] == 2019
t05_19 = t05_19_21_merg[msk19==True]
t05_19.reset_index(drop =True, inplace = True)
    #2020
msk20 = t05_19_21_merg["Periodo"] == 2020
t05_20 = t05_19_21_merg[msk20==True]
t05_20.reset_index(drop =True, inplace = True)
    #2021
msk21 = t05_19_21_merg["Periodo"] == 2021
t05_21 = t05_19_21_merg[msk21==True]
t05_21.reset_index(drop =True, inplace = True)
```

#### Imputar 2019 y 2020

##### Variables poblacionales

```python
###cargue de totales 
#cargue totales 2019-2022 
names_grl = ["Periodo", "Trimestre",  "Ocupados_T"]
t18_Tot = pd.read_excel("G:/Unidades compartidas/Mercado laboral/Datos/4_Output/ml_pob_impu_19_22.xlsx",     
                       sheet_name = "bogotá trim movil", header = 0,
                         nrows = 40, usecols = "A,B,L", names = names_grl)
##datos 2021
msk_T = t18_Tot["Periodo"]==2019
t18_19_Tot = t18_Tot[msk_T==True]
t18_19_Tot.reset_index(drop =True, inplace=True)
t18_19_Tot = t18_19_Tot.iloc[0:12:3]
t18_19_Tot.reset_index(drop =True, inplace=True)
##datos 2022
msk_T_2 = t18_Tot["Periodo"]==2020
t18_20_Tot = t18_Tot[msk_T_2 ==True]
t18_20_Tot.reset_index(drop =True, inplace=True)
t18_20_Tot = t18_20_Tot.iloc[0:12:3]
t18_20_Tot.reset_index(drop =True, inplace=True)
```

```python
#variables 
l1 = ['Unipersonal', 'Entre 2 y 10 personas',
       'Entre 11 y 50 personas', 'Más de 50 personas']
```

```python
#print(t05_20.iloc[1,2] + t05_20.iloc[1,3] + t05_20.iloc[1,4] + t05_20.iloc[1,5] + t05_20.iloc[1,6])
```

```python
#crear variable total ocuados en las 4 tablas
fram = [t05_19, t05_20, t18_21, t18_22]
for i in (fram):
    i["Ocupados"] = np.sum(i.iloc[:,2:7], axis=1)
```

```python
#imputar nan quedan nan
#x_19_imp= (x_19/(ocupados_19)) * ocu_t_imp_19
l1 = l1 + ["Ocupados"]
t05_19_imp = t05_19.copy()
t05_20_imp = t05_20.copy()

for i in (l1):
    t05_19_imp[i] = ((t05_19[i] / (t05_19["Ocupados"])) * (t18_19_Tot["Ocupados_T"])).astype(np.double)
    t05_20_imp[i] = ((t05_19_imp[i] *
                              (1 + ((t05_20[i] - t05_19[i]) / t05_19[i])))).astype(np.double)     
```

#### Concatenar

```python
#concat 2019-2022
f_frames = [t05_19_imp, t05_20_imp, t18_21, t18_22]
t05_18_19_22 = pd.concat(f_frames)
t05_18_19_22.reset_index(drop =True, inplace = True)
```

```python
### marca valores imputados
def m_imp(df):
    if ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.notnull(df["Unipersonal"]))): 
        x = 1
    elif ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.isnull(df["Unipersonal"]))):
        x = np.nan
    else:
        x = 0
    return(x)
```

```python
t05_18_19_22["valor Imputado"] = t05_18_19_22.apply(m_imp, axis=1)
```
