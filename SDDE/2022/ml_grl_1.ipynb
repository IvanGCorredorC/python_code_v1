#### Imputar 2019 y 2020

##### Variables poblacionales

```python
t05_19t.columns
```

```python
#variables
l1 = ['Población en edad de trabajar', 'Fuerza de trabajo',
       'Ocupados', 'Desocupados', 'Población fuera de la fuerza laboral',
       'Subocupados', 'Informales']
```

```python
t18_21t.columns
```

```python
### 2019_impu  = (2021 marco 2018 / 2021 marco 2005) * 2019 marco 2005
#impuar valores 2019. nan quedan nan
t05_19t_imp = t05_19t.copy()
t05_20t_imp = t05_20t.copy()
for i in (l1):
    t05_19t_imp[i] = (t18_21t[i] / t05_21t[i] * t05_19t[i]).astype(np.double)
    t05_20t_imp[i] = ((t05_19t_imp[i] * (1 + ((t05_20t[i] - t05_19t[i]) / t05_19t[i])))).astype(np.double)
```

##### Tasas

```python
t05_19t_imp.columns
```

```python
#variables
l2 = ["% población en edad de trabajar ", "TGP","TO","TD", "TS", "TI"]
l3_n = ["Población en edad de trabajar", "Fuerza de trabajo","Ocupados",
        "Desocupados", "Subocupados", "Informales"]
l4_d = ["Población total", "Población en edad de trabajar","Población en edad de trabajar",
        "Fuerza de trabajo", "Ocupados", "Ocupados"]
```

```python
### 2019_impu  = (2021 marco 2018 / 2021 marco 2005) * 2019 marco 2005
#impuar valores 2019. nan quedan nan
for i, j, k in zip(l2, l3_n, l4_d):
    t05_19t_imp[i] = (t05_19t_imp[j] / t05_19t_imp[k]*100).astype(np.double)
    t05_20t_imp[i] = (t05_20t_imp[j] / t05_20t_imp[k]*100).astype(np.double)
```

```python
t05_19t_imp = t05_19t_imp.reindex(columns=['Periodo', 'Trimestre', '% población en edad de trabajar ', 
                                           'TGP', 'TO', 'TD','TS','TI', 'Población total',
                                           'Población en edad de trabajar','Fuerza de trabajo',
                                           'Ocupados', 'Desocupados',
                                           'Población fuera de la fuerza laboral', 'Subocupados', 
                                           'Informales'])
t05_20t_imp = t05_20t_imp.reindex(columns=['Periodo', 'Trimestre', '% población en edad de trabajar ', 
                                           'TGP', 'TO', 'TD','TS', 'TI','Población total',
                                           'Población en edad de trabajar','Fuerza de trabajo',
                                           'Ocupados', 'Desocupados',
                                           'Población fuera de la fuerza laboral', 'Subocupados', 
                                           'Informales'])
```

#### Concatenar

```python
lframe = [t05_19t_imp, t05_20t_imp, t18_21t, t18_22t]
```

```python
ml_pob_impu_19_22 = pd.concat(lframe, axis = 0, join = 'outer', sort = False)
```

```python
#poblaciones a millones enteros
l_5 = ['Población total', 'Población en edad de trabajar', 'Fuerza de trabajo',
       'Ocupados', 'Desocupados', 'Población fuera de la fuerza laboral',
       'Subocupados', 'Informales']
for i in (l_5):
    ml_pob_impu_19_22[i] = ml_pob_impu_19_22[i]*1000 #np.around(ml_pob_impu_19_22[i], 0)
```

```python
### marca valores imputados
def m_imp(df):
    if ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.notnull(df["Población en edad de trabajar"]))): 
        x = 1
    elif ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.isnull(df["Población en edad de trabajar"]))):
        x = np.nan
    else:
        x = 0
    return(x)
```

```python
ml_pob_impu_19_22["valor Imputado"] = ml_pob_impu_19_22.apply(m_imp, axis=1)
```
