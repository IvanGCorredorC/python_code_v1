#### Imputar 2019 y 2020

##### Variables poblacionales

```python
###cargue de totales 
#cargue totales 2019-2022 
names_col_21 = ["Periodo", "Trimestre", "PET_T", "FT_T", "Ocupados_T", "Desocupados_T"]
t18_Tot = pd.read_excel("G:/Unidades compartidas/Mercado laboral/Datos/4_Output/ml_pob_impu_19_22.xlsx",     
                       sheet_name = "bogotá trim movil", header = 0,
                         nrows = 41, usecols = "A,B,J:M", names = names_col_21 )
##datos 2021
msk_T = t18_Tot["Periodo"]==2019
t18_19_Tot = t18_Tot[msk_T==True]
t18_19_Tot.reset_index(inplace=True)
##datos 2022
msk_T_2 = t18_Tot["Periodo"]==2020
t18_20_Tot = t18_Tot[msk_T_2 ==True]
t18_20_Tot.reset_index(inplace=True)
```

```python
#print(t18_21_sexo.columns, t05_21t.columns)
```

```python
#variables poblacionales
l1 = ['PET_h', 'FT_h', 'Desocupados_h', 'Ocupados_h', 'PET_m', 
      'FT_m', 'Desocupados_m', 'Ocupados_m']
ld = ['PET_m', 'FT_m', 'Desocupados_m', 'Ocupados_m', 'PET_h', 
      'FT_h', 'Desocupados_h', 'Ocupados_h']
lt = ["PET_T", "FT_T",  "Desocupados_T", "Ocupados_T",
      "PET_T", "FT_T", "Desocupados_T", "Ocupados_T"]
```

```python
### = ocupadasmujeresmarco2005proy2018 / (ocupadasmujeresmarco2005proy2018 + 
#ocupadoshombresmarco2005proy2018) * totalocupadosimputad
#impuar valores 2019. nan quedan nan
#h_19_imp= (h_19/(m_19+h_19)) * ocu_t_imp_19
t05_19t_imp = t05_19t.copy()
t05_20t_imp = t05_20t.copy()

for i, j, k in zip(l1,ld,lt):
    t05_19t_imp[i] = ((t05_19t[i] / (t05_19t[i]+ t05_19t[j])) * (t18_19_Tot[k]/1000)).astype(np.double)
    t05_20t_imp [i] = ((t05_19t_imp[i] *
                              (1 + ((t05_20t[i] - t05_19t[i]) / t05_19t[i])))).astype(np.double)      
```

```python
#variables
l2 = ['TGP_h','TGP_m', 'TO_h', 'TD_h', 'TO_m', 'TD_m']
l3_n = ['FT_h', 'FT_m', 'Ocupados_h', 'Desocupados_h', 'Ocupados_m', 'Desocupados_m']
l4_d = [ 'PET_h', 'PET_m', 'PET_h', 'FT_h', 'PET_m', 'FT_m']
```

```python
### 2019_impu  = (2021 marco 2018 / 2021 marco 2005) * 2019 marco 2005
#impuar valores 2019. nan quedan nan
for i, j, k in zip(l2, l3_n, l4_d):
    t05_19t_imp[i] = (t05_19t_imp[j] / t05_19t_imp[k]*100).astype(np.double)
    t05_20t_imp[i] = (t05_20t_imp[j] / t05_20t_imp[k]*100).astype(np.double)#.astype(int)
```

#### Concatenar

```python
#concat 2019-2022
f_frames = [t05_19t_imp, t05_20t_imp, t18_21_sexo, t18_22_sexo]
desag_impu_19_22 = pd.concat(f_frames)
```

```python
##eliminar variables que no tienen marco muestral 2005
desag_impu_19_22.drop(columns=["%PET_h", "Población total_hombre", "%PET_m", "Población total_mujer"], inplace = True)
```

```python
#poblaciones a millones enteros de l1
for i in (l1):
    desag_impu_19_22[i] = desag_impu_19_22[i]*1000 #np.around(desag_impu_19_22[i], 0)
```

```python
### marca valores imputados
def m_imp(df):
    if ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.notnull(df["PET_h"]))): 
        x = 1
    elif ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.isnull(df["PET_h"]))):
        x = np.nan
    else:
        x = 0
    return(x)
```

```python
desag_impu_19_22["valor Imputado"] = desag_impu_19_22.apply(m_imp, axis=1)
```
