#### Transponer

```python
#establecer indice y transponer
frames_18 = [t18_21_j, t18_22_j]

for i in (frames_18):
    i.set_index("Poblaciones", inplace =True)

t18_21_j_t = t18_21_j.transpose()
t18_22_j_t = t18_22_j.transpose()
```

```python
## homogenizar nombres de variables y asignar variable de año
frames_18_t = [t18_21_j_t, t18_22_j_t]

names_j = {'index':'Trimestre','% población en edad de trabajar ': "%PET_j", 
           '% Fuera de la fuerza laboral / PET 15 a 28 años': '%FFL_j/PET_j',
           'Población en edad de trabajar':'PET_T',
           'Población en edad de trabajar 15 a 28 años':'PET_j', 
           'Fuerza de trabajo':'FT_j', 'Ocupados': 'Ocupados_j',
           'Desocupados': 'Desocupados_j', 'Fuera de la fuerza laboral': 'FFL_j',
           'TGP':'TGP_j', 'TO': 'TO_j', 'TD':'TD_j'}
# names_nj = {'index':'Trimestre','% población en edad de trabajar ': "%PET_nj", 
#                               'Población en edad de trabajar':'PET_nj', 
#                                'Fuerza de trabajo':'FT_nj', 'Desocupados': 'Desocupados_nj',
#                               'Ocupados': 'Ocupados_nj', 'TGP':'TGP_nj', 'TO': 'TO_nj', 'TD':'TD_nj'}
names_col = [names_j, names_j]
per = [2021, 2022]

for i, j, k in zip(frames_18_t, names_col, per):
    i.reset_index(inplace =True)
    i.rename(columns = j, inplace =True)
    i = i.insert(0, "Periodo",  k)
```

```python
#cargue totales 2019-2022 
names_col_21 = ["Periodo", "Trimestre", 'TGP', 'TO', 'TD',"PET_T", 
                "FT_T", "Ocupados_T", "Desocupados_T", "FFL_T"]
t18_Tot = pd.read_excel("G:/Unidades compartidas/Mercado laboral/Datos/4_Output/ml_pob_impu_19_22.xlsx",     
                       sheet_name = "bogotá trim movil", header = 0,
                         nrows = 41, usecols = "A,B, D:F, J:N", names = names_col_21 )
##datos 2021
msk_19_T = t18_Tot["Periodo"]==2019
t18_19_Tot = t18_Tot[msk_19_T==True]
t18_19_Tot.reset_index(inplace=True)

msk_T = t18_Tot["Periodo"]==2021
t18_21_Tot = t18_Tot[msk_T==True]
t18_21_Tot.reset_index(inplace=True)
##datos 2022
msk_T_2 = t18_Tot["Periodo"]==2022
t18_22_Tot = t18_Tot[msk_T_2 ==True]
t18_22_Tot.reset_index(inplace=True)
```

```python
t18_21_Tot.columns
```

```python
### crear variables para los no jovenes 2021-2022 marco 2018 dividios en mil
# ['PET', 'FT', 'Ocupados', 'Desocupados'] 2021 - ['PET_j', 'FT_j', 'Ocupados_j', 'Desocupados_j'] 2021
new_nj = ['PET_nj', 'FT_nj', 'Ocupados_nj', 'Desocupados_nj']
var_T = ["PET_T", "FT_T", "Ocupados_T", "Desocupados_T"]
var_j = ['PET_j', 'FT_j', 'Ocupados_j', 'Desocupados_j']
for i,j, k in zip(new_nj,var_T, var_j):
    t18_21_j_t[i] = ((t18_21_Tot[j]/1000) - t18_21_j_t[k])
    t18_22_j_t[i] = ((t18_22_Tot[j]/1000) - t18_22_j_t[k])
```

```python
#variables poblacionales , 'TGP_nj',	'TO_nj','TD_nj'
l11 = ['TGP_nj','TO_nj','TD_nj']
l33_n = ['FT_nj', 'Ocupados_nj', 'Desocupados_nj']
l44_d = ['PET_nj', 'PET_nj', 'FT_nj']
for i, j, k in zip(l11, l33_n, l44_d):
    t18_21_j_t[i] = (t18_21_j_t[j] / t18_21_j_t[k]*100).astype(np.double)
    t18_22_j_t[i] = (t18_22_j_t[j] / t18_22_j_t[k]*100).astype(np.double)
```

```python
#variables poblacionales
l1 = ['PET_j', 'FT_j', 'Ocupados_j', 'Desocupados_j', 
      'PET_nj', 'FT_nj', 'Ocupados_nj', 'Desocupados_nj']
ld = ['PET_nj', 'FT_nj', 'Ocupados_nj', 'Desocupados_nj',
      'PET_j', 'FT_j', 'Ocupados_j', 'Desocupados_j']
lt = ["PET_T", "FT_T",  "Ocupados_T", "Desocupados_T",
      "PET_T", "FT_T",  "Ocupados_T", "Desocupados_T"]
```

```python
### 2019_impu  = (2021 marco 2018 / 2021 marco 2005) * 2019 marco 2005
#impuar valores 2019. nan quedan nan
t05_19t_imp = t05_19t.copy()
t05_20t_imp = t05_20t.copy()

for i, j, k in zip(l1, ld, lt):
    t05_19t_imp[i] = ((t05_19t[i] / (t05_19t[i]+ t05_19t[j])) * (t18_19_Tot[k]/1000)).astype(np.double)
    t05_20t_imp [i] = ((t05_19t_imp[i] *
                              (1 + ((t05_20t[i] - t05_19t[i]) / t05_19t[i])))).astype(np.double)      
```

```python
#impuar valores 2019. nan quedan nan
for i, j, k in zip(l2, l3_n, l4_d):
    t05_19t_imp[i] = (t05_19t_imp[j] / t05_19t_imp[k]*100).astype(np.double)
    t05_20t_imp[i] = (t05_20t_imp[j] / t05_20t_imp[k]*100).astype(np.double)#.astype(int)
```

#### Concatenar

```python
#concat 2021-2022 hombres y mujeres.
f_frames = [t05_19t_imp, t05_20t_imp, t18_21_j_t, t18_22_j_t]
j_nj_impu_19_22 = pd.concat(f_frames)
```

```python
##eliminar variables que no tienen marco muestral 2005
j_nj_impu_19_22.drop(columns=["%PET_j",	"%FFL_j/PET_j",	"PET_T", "FFL_j"], inplace = True)
```

```python
#poblaciones a millones enteros de l1
for i in (l1):
    j_nj_impu_19_22[i] = j_nj_impu_19_22[i]*1000
```

```python
### marca valores imputados
def m_imp(df):
    if ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.notnull(df["PET_j"]))): 
        x = 1
    elif ((df["Periodo"] == 2019 or df["Periodo"] == 2020) 
        and (pd.isnull(df["PET_j"]))):
        x = np.nan
    else:
        x = 0
    return(x)
```

```python
j_nj_impu_19_22["valor Imputado"] = j_nj_impu_19_22.apply(m_imp, axis=1)
```
