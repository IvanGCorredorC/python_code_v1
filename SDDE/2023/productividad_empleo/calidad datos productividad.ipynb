```python
# filtro activas y renovadas cr_can_re_19_22['estado_matricula']=="MA") & 
renovadas = cr_can_re_19_22[(cr_can_re_19_22['estado_matricula']=="MA") &
                            (cr_can_re_19_22["tipo empresa"]=="renovadas")].reset_index(drop=True)
#filtro variables
renovadas = renovadas[['num_matricula', 'vigencia', 'fec_matricula', 'año_matricula',
       'fec_vigencia', 'fec_cancelacion', 'año_cancelacion', 'fec_renova', 'ano_renova',
       'tipo_persona', 'cantidad_establecimientos', 'organizacion_juridica', 'rep_legal',
       'tipo_id_replegal', 'estado_matricula', 'resultado_periodo',
       'utilidad_perdida_operacional', 'activo_total', 'pasivo_total',
       'patrimonio_neto', 'tamano_activos', 'tamano_ingresos',
       'numero_empleados', 'importador_exportador', 'ciiu_1', 'sectores_pib',
       'sectores_14', 'macro_sec',  'sexo', 'año', 'nacionalidad','ROA', 'ROE']]
#borrar caché
del  cr_can_re_19_22
```

```python
# filtro naturales
renovadas_n = renovadas[renovadas['tipo_persona']== 1]
# filtro juridicas
renovadas_j = renovadas[renovadas['tipo_persona']== 2]
```

##### <span style="color:yellow">Completa, natural, jurídica</span>

```python
#estadístcas descriptivas
# tablas
nom_t1 = [renovadas, renovadas_n, renovadas_j]
# listas para almacenar información
nom_t2 = ["completa", "natural", "juridica"]
# variables
var_clave = 'utilidad_perdida_operacional'
var1 = ["mediana", "ceros", "% ceros", "dis ceros", "% dis ceros", "nulos"]

# procedimiento
for i, k in zip(nom_t1, nom_t2):
# crear variables en la tabla
    df = i[[var_clave]].describe().transpose()
    # crear variables adicionales en cada tabla
    for j in var1:
        if j == 'mediana':
            df[j] = i[[var_clave]].median()
        elif j == 'ceros':
            df[j] = (np.count_nonzero(i[var_clave] == 0.0))
        elif j == '% ceros':
            df[j] = ((np.count_nonzero(i[var_clave] == 0.0) / len(i[var_clave])) * 100 )
        elif j == 'dis ceros':
            df[j] = (np.count_nonzero(i[var_clave] != 0.0))
        elif j == '% dis ceros':
            df[j] = ((np.count_nonzero(i[var_clave] != 0.0) / len(i[var_clave])) * 100)
        elif j == 'nulos':
            df[j] = i[[var_clave]].isna().sum()
    # crear objetos con las tablas y reindexarlas
    exec(f"{k} = df.reset_index(drop=True).rename(index=dict([(0, k)]))")
# unir tablas
t1 = pd.concat([completa, natural, juridica], axis=0)
t1
```

##### <span style="color:yellow">Tamaño, sector, sexo, año, organización jurídica </span>

```python
# Estadísticos agrupados por una sola variable: tamaño, sector, sexo, año, organización jurídica
# variables 
variables = ['tamano_activos','macro_sec','sexo','año', 'organizacion_juridica']

# listas para almacenar información
nom_tablas = ['tamaño1', 'sector1', 'sexo1', 'año1', 'org_jur1']
nom_tablas2 = ['tamaño2', 'sector2', 'sexo2', 'año2', 'org_jur2']
nom_salidas = ['tamaño', 'sector', 'sexo', 'año', 'org_jur']

# procedimiento
for i, j, k,l in zip(variables, nom_tablas, nom_tablas2, nom_salidas):
    #estadísticos generales
    df1 = renovadas_j.groupby([i])[var_clave].describe()
    #estadísticos adicionales
    df2 = renovadas_j.groupby([i]).apply(
        lambda x: pd.Series(
            {
                'mediana':x[var_clave].median(),
                             'ceros': np.count_nonzero(x[var_clave]==0.0),
                             '% ceros':((np.count_nonzero(x[var_clave]== 0.0) / 
                                          len(x[var_clave])) * 100 ),
                            'dis ceros':(np.count_nonzero(x[var_clave]!= 0.0)),
                             '% dis ceros':((np.count_nonzero(x[var_clave]!= 0.0)/  
                                              len(x[var_clave])) * 100),
                            'nulos':(x[var_clave].isna().sum())
                            }
        )
    )
    # salidas de información por cada vuelta de bucle
    exec(f"{j} = df1")
    exec(f"{k} = df2")
    #concatenar a la derecha (wide)
    exec(f"{l} = pd.merge({j}, {k}, on=i)")
    #imprimir salida
    exec(f"print({l})")
```

```python
# estadísticos agrupados por 2 o más variables
# tamaño y año
ta_año1 = renovadas_j.groupby(["tamano_activos", "año"])[var_clave].describe()
# otras varaibles
ta_año2 = renovadas_j.groupby(["tamano_activos", "año"]).apply(
    lambda x: pd.Series({"mediana": x[var_clave].median(), 
                         "ceros":(np.count_nonzero(x[var_clave] == 0.0)),
                         "% ceros":((np.count_nonzero(x[var_clave] == 0.0) / 
                                      len(x[var_clave])) * 100 ),
                         "dis ceros":(np.count_nonzero(x[var_clave]!= 0.0)),
                         "% dis ceros":((np.count_nonzero(x[var_clave]!= 0.0) /
                                          len(x[var_clave])) * 100),
                         "nulos":(x[var_clave].isna().sum())
        }))
### merge
ta_año = pd.merge(ta_año1, ta_año2 , on=["tamano_activos", "año"])
ta_año
```

```python
# tamaño para activo total, patrimonio_neto y resultado_periodo
vars_ad = ["activo_total", "patrimonio_neto", "resultado_periodo"]

# listas para almacenar información
tablas_ad = ['ta_act1', 'ta_pat1', 'res_ejer1']
tablas2_ad = ['ta_act2', 'ta_pat2', 'res_ejer2']
salidas_ad = ['ta_act', 'ta_pat','res_ejer']
#procedimiento
for i, j, k,l in zip(vars_ad, tablas_ad, tablas2_ad, salidas_ad):
    #estadísticos generales
    df1 = renovadas_j.groupby(["tamano_activos"])[i].describe()
    #estadísticos adicionales
    df2 = renovadas_j.groupby(["tamano_activos"]).apply(
        lambda x: pd.Series({"mediana": x[i].median(), 
                             "ceros":(np.count_nonzero(x[i] == 0.0)),
                             "% ceros":((np.count_nonzero(x[i] == 0.0) / 
                                          len(x[i])) * 100 ),
                             "dis ceros":(np.count_nonzero(x[i]!= 0.0)),
                             "% dis ceros":((np.count_nonzero(x[i]!= 0.0) /
                                              len(x[i])) * 100),
                             "nulos":(x[i].isna().sum())
        }))
    # salidas de información por cada vuelta de bucle
    exec(f"{j} = df1")
    exec(f"{k} = df2")
    #concatenar a la derecha (wide)
    exec(f"{l} = pd.merge({j}, {k}, on='tamano_activos')")
    #imprimir salida
    #exec(f"print({l})")
```

### 1.2 Comparativo EAM - CCB

```python
#cargue EAM
EAM_15_19 = pd.read_stata(p3)
# VALORVEN: Ingresos; PERTOTAL: Empleados, VALAGRI = Valor agregado, 
# c3r2c3 = salarios personal permanente, utilidad = VALAGRI - c3r2c3, ACTIVFI = Activo total,
# ROA = utilidad / ACTIVFI
```

```python
EAM_15_19["VALORVEN"], EAM_15_19["utilidad"], EAM_15_19["ACTIVFI"] = EAM_15_19["VALORVEN"]*1000, EAM_15_19["utilidad"]*1000, EAM_15_19["ACTIVFI"]*1000 
```

```python
# EAM industria
# variables
v_clave_EAM = ["utilidad", "ACTIVFI", "ROA"]
#tablas salida
salidas_EAM = ["EAM_utilidad","EAM_activo","EAM_ROA"]
#procedimiento
for i, k in zip(v_clave_EAM, salidas_EAM):
    # general
    df1 = EAM_15_19.groupby(["year"])[i].describe()
    # otras varaibles
    df2 = EAM_15_19.groupby(["year"]).apply(
        lambda x: pd.Series({"mediana": x[i].median(), 
                             "ceros":(np.count_nonzero(x[i] == 0.0)),
                             "% ceros":((np.count_nonzero(x[i] == 0.0) / 
                                          len(x[i])) * 100 ),
                             "dis ceros":(np.count_nonzero(x[i]!= 0.0)),
                             "% dis ceros":((np.count_nonzero(x[i]!= 0.0) /
                                              len(x[i])) * 100),
                             "nulos":(x[i].isna().sum())
            }))
    # merge
    exec(f'{k} = pd.merge(df1,  df2 , on=["year"])')
    # ordenar variables
    exec(f'{k} = {k}[["min","max", "mean", "mediana","std","25%", "50%", "75%","count", "ceros", "dis ceros", "% ceros",  "% dis ceros", "nulos"]]')
```
