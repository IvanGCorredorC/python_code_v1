```python
#agregar variable nombre a regiones ipm dpto
dpto_19_21["nombre_region"] = dpto_19_21["REGION"].replace([1,2,3,4,5,6,7,8,9], ["Caribe", "Oriental", "Central", "Pacífica (sin valle)", "Bogotá", 
                                                                       "Antioquia", "Valle del Cauca", 
                                                                        "San Andrés y Providencia", "Orinoquía y Amazonía"])
```

### Filtro para 2021

```python
## base pobreza mon y ext
vic_21 = vic_19_21[vic_19_21["periodo"]==2021]
# reset index
vic_21 = vic_21.reset_index(drop=True)
```

```python
# base ipm nal y dpto
## nal
nal_21  = nal_19_21[nal_19_21 ["periodo"]==2021]
# reset index
nal_21 = nal_21.reset_index(drop=True)
##dpto
dpto_21 = dpto_19_21[dpto_19_21["periodo"]==2021]
# reset index
dpto_21 = dpto_21.reset_index(drop=True)
```

### Departamento

```python
## víctimas de desplazamiento en pobreza monetaria por departamento 2021
tvic_monetaria_dpto = vic_21.groupby(["depto", "nombre_dpto"]).apply(
    lambda x: pd.Series({
        # vic tot pobreza monetaria
        '%_vic_hechos_pobr_mon': round((x['pobre'] * x['es_victima'] * x['fex_c_viv']).sum() / (x['es_victima'] * x['fex_c_viv']).sum() * 100,1),
        'vic_hechos_pobr_mon': round((x['pobre'] * x['es_victima'] * x['fex_c_viv']).sum(),0),
        
        # vic tot pobreza extrema
        '%_vic_hechos_pobr_ext': round((x['indigente'] * x['es_victima'] * x['fex_c_viv']).sum() / (x['es_victima'] * x['fex_c_viv']).sum() * 100,1),
        'vic_hechos_pobr_ext': round((x['indigente'] * x['es_victima'] * x['fex_c_viv']).sum(),0),
        
        # vic des pobreza monetaria
        '%_des_pobr_mon': round((x['pobre'] * x['es_desplazado'] * x['fex_c_viv']).sum() / (x['es_desplazado'] * x['fex_c_viv']).sum() * 100,1),
        'des_pobr_mon': round((x['pobre'] * x['es_desplazado'] * x['fex_c_viv']).sum(),0),
         
        # vic des pobreza extrema
        '%_des_pobr_ext': round((x['indigente'] * x['es_desplazado'] * x['fex_c_viv']).sum() / (x['es_desplazado'] * x['fex_c_viv']).sum() * 100,1),
        'des_pobr_ext': round((x['indigente'] * x['es_desplazado'] * x['fex_c_viv']).sum(),0)
    })
).reset_index()

tvic_monetaria_dpto.head(30)
```

```python
#pasar variables base nal a formato numero antes de agrupar
nal_21["es_desplazado"] = pd.to_numeric(nal_21["es_desplazado"], downcast="integer",errors="coerce")
nal_21["fex_c_des"] = pd.to_numeric(nal_21["fex_c_des"], downcast="integer",errors="coerce")
nal_21["fex_c_victi"] = pd.to_numeric(nal_21["fex_c_victi"], downcast="integer",errors="coerce") 
```

```python
### regiones IPM

tvic_ipm_reg = nal_21.groupby(["REGION"]).apply(
    lambda x: pd.Series({
        # vic tot ipm
        '%_vic_hechos_ipm': round((x['pobre'] * x['fex_c_victi']).sum() / (x['victima'] * x['fex_c_victi']).sum() * 100,1),
        'vic_hechos_ipm': round((x['pobre'] * x['fex_c_victi']).sum(),0),
        
        # vic des ipm
        '%_des_ipm': round((x['pobre'] * x['fex_c_des']).sum() / (x['es_desplazado'] * x['fex_c_des']).sum() * 100,1),
        'des_ipm': round((x['pobre']  * x['fex_c_des']).sum(),0)
         
    })
).reset_index()

tvic_ipm_reg.head(30)
```

```python
#pasar variables base dpto a formato numero antes de agrupar
dpto_21["es_desplazado"] = pd.to_numeric(dpto_21["es_desplazado"], downcast="integer",errors="coerce")
dpto_21["fex_c_des"] = pd.to_numeric(dpto_21["fex_c_des"], downcast="integer",errors="coerce")
dpto_21["fex_c_victi"] = pd.to_numeric(dpto_21["fex_c_victi"], downcast="integer",errors="coerce") 
```

```python
### departamento ipm

tvic_ipm_dpto = dpto_21.groupby(["nombre_region", "DEPARTAMENTO", "nombre_departamento" ]).apply(
    lambda x: pd.Series({
        # vic tot ipm
        '%_vic_hechos_ipm': round((x['pobre'] * x['fex_c_victi']).sum() / (x['victima'] * x['fex_c_victi']).sum() * 100,1),
        'vic_hechos_ipm': round((x['pobre'] * x['fex_c_victi']).sum(),0),
        
        # vic des ipm
        '%_des_ipm': round((x['pobre'] * x['fex_c_des']).sum() / (x['es_desplazado'] * x['fex_c_des']).sum() * 100,1),
        'des_ipm': round((x['pobre']  * x['fex_c_des']).sum(),0)
         
    })
).reset_index()

tvic_ipm_dpto.head(30)
```
