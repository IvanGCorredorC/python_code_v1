### Estandarizar

```python
## borrar columnas innecesarios
nal_19_22.drop(columns=["Unnamed: 0"], inplace =True)
dpto_19_22.drop(columns=["Unnamed: 0"], inplace =True)
```

```python
### estandarizar variable dpto
vic_19_22["dpto"] = pd.to_numeric(vic_19_22["dpto"], downcast = 'integer')
vic_19_22["depto"] = pd.to_numeric(vic_19_22["depto"], downcast = 'integer')
```

```python
#agregar variable nombres de departamentos
def nom_dpto(df):
    cod_d = [5, 8, 11, 13, 15, 17, 18, 19, 20, 23, 25, 27, 41, 44, 47, 50, 52, 54, 63, 66, 68,
             70, 73, 76, 81, 85, 86, 88, 91, 94, 95, 97, 99]
    dep = ["Antioquia", "Atlántico", "Bogotá", "Bolívar", "Boyacá",
           "Caldas", "Caquetá", "Cauca", "Cesar", "Córdoba", "Cundinamarca",
           "Chocó", "Huila", "La Guajira", "Magdalena", "Meta", "Nariño",
           "Norte de Santander", "Quindio", "Risaralda", "Santander",
           "Sucre", "Tolima", "Valle del Cauca", "Arauca", "Casanare",
           "Putumayo", "Providencia y Santa Catalina", "Amazonas",
           "Guainía", "Guaviare", "Vaupés", "Vichada"]
    
    x = np.nan  # Initialize x outside the loop

    for i, j in zip(cod_d, dep):
        if df["dpto"] == i or df["depto"] == i:
            x = j
            break  # Break out of the loop once a match is found

    return x

vic_19_22["nombre_dpto"] = vic_19_22.apply(nom_dpto, axis=1)
```

```python
#agregar variable código ciudades capítales
vic_19_22["cod_dane"]= vic_19_22["dominio"].replace(['ARMENIA','BARRANQUILLA','BOGOTA','BUCARAMANGA','CALI','CARTAGENA',
                                                     'CUCUTA', 'FLORENCIA','IBAGUE','MANIZALES','MEDELLIN','MONTERIA',
                                                     'NEIVA','PASTO','PEREIRA','POPAYAN','QUIBDO','RIOHACHA',
                                                     'SANTA MARTA','SINCELEJO','TUNJA','VALLEDUPAR','VILLAVICENCIO'],
                                                                [63001, 8001, 11001, 68001, 6001, 13001,
                                                                 54001, 18001, 73001, 17001, 5001, 23001,
                                                                 41001, 52001, 66001, 19001, 27001, 44001,
                                                                 47001, 70001, 15001, 20001, 50001])
```

```python
#agregar variable nombre a regiones ipm dpto
dpto_19_22["nombre_region"] = dpto_19_22["REGION"].replace([1,2,3,4,5,6,7,8,9], 
                                                           ["Caribe", "Oriental", "Central", 
                                                             "Pacífica (sin valle)", "Bogotá", 
                                                             "Antioquia", "Valle del Cauca", 
                                                             "San Andrés y Providencia", "Orinoquía y Amazonía"])
```

```python
#agregar variable nombre a departamento ipm dpto
dpto_19_22["nombre_departamento"] = dpto_19_22["DEPARTAMENTO"].replace([5,8,11,13,15,17,18,19,20,23,25,27,41,44,47,50,52,54,63,66,68,
                                                                70, 73, 76, 81, 85,86, 88, 91,94, 95,97,99],
                                                                ["Antioquia", "Atlántico", "Bogotá", "Bolívar","Boyacá", 
                                                                 "Caldas","Caquetá","Cauca", "Cesar","Córdoba", "Cundinamarca",
                                                                "Chocó", "Huila", "La Guajira", "Magdalena", "Meta", "Nariño",
                                                                "Norte de Santander", "Quindio", "Risaralda", "Santander",
                                                                "Sucre", "Tolima", "Valle del Cauca", "Arauca", "Casanare",
                                                                "Putumayo", "Providencia y Santa Catalina", "Amazonas", 
                                                                "Guainía", "Guaviare", "Vaupés", "Vichada"])
```

```python
## víctimas de desplazamiento en pobreza monetaria por departamento 2022
pm_dpto22 = vic_22.groupby(["dpto", "nombre_dpto"]).apply(
    lambda x: pd.Series({
        # vic tot pobreza monetaria
        '%_vic_hechos_pobr_mon': round((x['pobre'] * x['es_victima'] * x['fex_c_viv']).sum() / (x['es_victima'] * x['fex_c_viv']).sum() * 100,1),
        'vic_hechos_pobr_mon': round((x['pobre'] * x['es_victima'] * x['fex_c_viv']).sum(),0),
        
        # vic tot pobreza extrema
        '%_vic_hechos_pobr_ext': round((x['indigente'] * x['es_victima'] * x['fex_c_viv']).sum() / (x['es_victima'] * x['fex_c_viv']).sum() * 100,1),
        'vic_hechos_pobr_ext': round((x['indigente'] * x['es_victima'] * x['fex_c_viv']).sum(),0),
        
        # vic des pobreza monetaria
        '%_des_pobr_mon': round((x['pobre'] * x['es_desplazado'] * x['fex_c_viv']).sum() / (x['es_desplazado'] * x['fex_c_viv']).sum() * 100,1),
        'des_pobr_mon': round((x['pobre'] * x['es_desplazado'] * x['fex_c_viv']).sum(),0),
         
        # vic des pobreza extrema
        '%_des_pobr_ext': round((x['indigente'] * x['es_desplazado'] * x['fex_c_viv']).sum() / (x['es_desplazado'] * x['fex_c_viv']).sum() * 100,1),
        'des_pobr_ext': round((x['indigente'] * x['es_desplazado'] * x['fex_c_viv']).sum(),0)
    })
).reset_index()

pm_dpto22.head(33)
```

```python
ipm_reg22a = nal_22.groupby(["REGION"]).apply(
    lambda x: pd.Series({
        # vic tot ipm
        '%_vic_hechos_ipm': round((x['pobre'] * x['fex_c_victi']).sum() / (x['es_victima'] * x['fex_c_victi']).sum() * 100, 1),
        'vic_hechos_ipm': round((x['pobre'] * x['fex_c_victi']).sum(), 0)})
).reset_index()

ipm_reg22b = nal_22[nal_22["es_desplazado"]==1].groupby(["REGION"]).apply(
    lambda x: pd.Series({
        # vic des ipm
        '%_des_ipm': round((x['pobre'] * x['fex_c_victi']).sum() / (x['es_desplazado'] * x['fex_c_victi']).sum() * 100, 1),
        'des_ipm': round((x['pobre'] * x['es_desplazado'] * x['fex_c_victi']).sum(), 0)
    })
).reset_index()
  
ipm_reg22 = pd.merge(ipm_reg22a, ipm_reg22b, on =["REGION"])
ipm_reg22.head(10)
```
