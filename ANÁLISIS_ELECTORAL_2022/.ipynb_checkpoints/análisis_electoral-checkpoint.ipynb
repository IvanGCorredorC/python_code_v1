```python
# reindexar filas y columnas
votos_munir = votos_muni.reset_index(col_level = 1)
votos_munir.columns = votos_munir.columns.droplevel(0)
votos_munir.columns.name = None
```

```python
# agrupar
cons_f= votos_munir.groupby(
    ["COD_DANE","DEPTO","MUN","COD_ZONA", "COD_PUESTO"]).agg(
    {"COD_MESA": np.size, "FRANCIA MÁRQUEZ":np.sum}).reset_index()
```

```python
#cargar censo electoral
censo_electoral = pd.read_excel("D:/one_drive_unal/OneDrive - Universidad Nacional de Colombia/análisis_electoral_2022/Divipole_Elecciones_Congreso_13_03_2022.xlsx",
                           sheet_name="divipol", header=0)
```

```python
censo_electoral = censo_electoral.iloc[:12512,:]
```

#### Estandarizar llaves para pegue

```python
##set(cons_f["COD_PUESTO"])
cons_f["COD_PUESTO"] = cons_f["COD_PUESTO"].astype(str).str.zfill(2)
```

```python
cons_f["COD_ZONA"] = cons_f["COD_ZONA"].astype(str)
```

```python
censo_electoral["zz"] = censo_electoral["zz"].astype(int)
```

```python
censo_electoral["zz"] = censo_electoral["zz"].astype(str)
```

```python
#normalizar a mayuscula y sin acentos depto
import unicodedata
def normalize_string(string):
    string = unicodedata.normalize('NFKD', string).encode('ASCII', 'ignore').decode('utf-8')
    string = string.replace('.', '') # remove periods from string
    return string.upper()
```

```python
cons_f['DEPTO'] = cons_f['DEPTO'].apply(normalize_string)
```

```python
#instalar paquete fuzy
#pip install fuzzywuzzy
```

```python
#importar 
from fuzzywuzzy import fuzz, process
```

```python
# function to find the closest match for a string in a list of strings
def get_closest_match(string, choices):
    return process.extractOne(string, choices, scorer=fuzz.token_sort_ratio)[0]

# create a dictionary to map the values in cons_f to the values in censo_electoral using fuzzy matching
# depto_map = {}
# for depto in cons_f['DEPTO'].unique():
#     closest_match = get_closest_match(depto, censo_electoral['departamento'].unique())
#     depto_map[depto] = closest_match

depto_map = {}
for depto in censo_electoral['departamento'].unique():
    closest_match = get_closest_match(depto, cons_f['DEPTO'].unique())
    if closest_match in cons_f['DEPTO'].unique():
        depto_map[depto] = closest_match
#censo_electoral['departamento'] = censo_electoral['departamento'].replace(depto_map)
```

```python
depto_map['TOLIMA'] = 'TOLIMA'
depto_map['VALLE'] = 'VALLE'
depto_map['CONSULADOS'] = 'CONSULADOS'
```

```python
#replace the values in censo_electoral  with the values in cons_f
censo_electoral['departamento'] = censo_electoral['departamento'].apply(lambda x: depto_map[x] if x in depto_map else x)
```

```python
#normalizar a mayuscula y sin acentos municipio en cons_f
cons_f["MUN"] = cons_f["MUN"].apply(normalize_string)
```

```python
#normalizar a mayuscula y sin acentos municipio en censo_electoral["municipio"]
censo_electoral["municipio"]= censo_electoral["municipio"].apply(normalize_string)
```

```python
cons_censo = pd.merge(left=cons_f, right=censo_electoral, 
                            left_on=["DEPTO","MUN","COD_ZONA", "COD_PUESTO"], 
                            right_on=["departamento", "municipio",  "zz", "pp"], how = 'outer', sort =False,
                            validate = '1:m', suffixes=("_resultado", "_censo"),
                            indicator = 'indicator_column')
```

```python
res_bog_f = cons_censo[cons_censo["municipio"]== "BOGOTA DC"] ## filtro bogotá
```

```python
# lista votos por localidad ordenados
res_bog_fg = res_bog_f.groupby("comuna").aggregate({"FRANCIA MÁRQUEZ":np.sum}).sort_values(by= "FRANCIA MÁRQUEZ", ascending = False)
```
